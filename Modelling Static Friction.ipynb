# Modelling Static Friction

### Procedure

Weigh an object and place it on the table. Using a spring scale, pull the box and measure the force necessary to move it. Place weights, one at a time in increasing order, and measure the force used to overcome static friction each time. Repeat the experiment on an incline.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def model (x ,a, b):
    return a*x + b
xm = np.linspace(0, 700, 400)
x=[0, 100, 200, 300, 400, 500]
y=[.5, .7, 1.25, 1.75, 1.9, 2.2]

plt.plot(x,y,'b.', label="Flat")

a,b = curve_fit(model, x, y)[0]
print(a,b)
plt.plot(xm, model(xm, a, b))
plt.xlabel("Mass (g)")
plt.ylabel("Force Applied (N)")

def model (x ,a, b):
    return a*x + b
xm = np.linspace(0, 700, 400)
x=[0, 100, 200, 300, 400, 500]
y=[1.05, 1.36, 1.626, 2.489, 3.251, 3.615]

plt.plot(x,y,'g.', label="17.5° incline")

a,b = curve_fit(model, x, y)[0]
print(a,b)
plt.plot(xm, model(xm, a, b))
plt.xlabel("Mass (g)")
plt.ylabel("Frictional Force (N)")

def model (x ,a, b):
    return a*x + b
xm = np.linspace(0, 700, 400)
x=[0, 100, 200, 300, 400, 500]
y=[.75, 1.26, 1.726, 2.489, 3.052, 3.215]



plt.plot(x,y,'r.', label="14° Incline")
a,b = curve_fit(model, x, y)[0]
print(a,b)
plt.plot(xm, model(xm, a, b))
plt.xlabel("Mass (g)")
plt.ylabel("Frictional Force (N)")
plt.legend(loc=2)
plt.show()

def model (x ,a, b):
    return a*x + b
xm = np.linspace(0, 4, 10)
x=[0, 0.95, 1.9, 2.85, 3.8, 4.75]
y=[.75, 1.26, 1.72, 2.48, 3.05, 3.21]
plt.plot(x,y,'b.', label="Flat")

a,b = curve_fit(model, x, y)[0]
print(a,b)
plt.plot(xm, model(xm, a, b))
plt.xlabel("Frictional Force")
plt.ylabel("Normal Force")
plt.title("Normal Force vs Frictional Force")
plt.show()
```

### Results

Linear equation for frictional force on an incline: 

$$ f_s = F_{sp} - mg (sin{\theta}) $$ 


### Conclusion

Through our experiments, we found that there is a linear relationship between the mass being moved and the force required to overcome static friction. Unfortunately, our model could not be made more accurate due to imperfect conditions, an inconsistent spring scale, and possibly human error. When placed on an incline, a greater amount of force is required to overcome static friction. The higher the incline, the higher the frictional force. Our data shows that on an incline there is a lot more friction than on a flat surface.
